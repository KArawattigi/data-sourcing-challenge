```python
# Extract the title from the "headline.main" column and
# save it to a new column "title"
# Title is between unicode characters \u2018 and \u2019. 
# End string should include " Review" to avoid cutting title early
 
def get_title(col_value):
    ar1 = str(col_value).split("Review")
    if ar1[0]!='': 
        return ar1[0].strip().strip('\u2018').strip('\u2019')
    else:
        #if title is in review 
        #find last '\u2019' and strip every thing after that 
        str_t = ar1[1].strip(': \u2018')
        ar2=str_t.split('\u2019')
        return str_t.strip(ar2[-1]).strip('\u2019')
    
    

df_reviews_List['title'] = df_reviews_List['headline.main'].apply(get_title)
df_reviews_List
```

```python
# Extract 'name' and 'value' from items in "keywords" column
def extract_keywords(keyword_list):
    extracted_keywords = ""
    for item in keyword_list:
        # Extract 'name' and 'value'
        keyword = f"{item['name']}: {item['value']};" 
        # Append the keyword item to the extracted_keywords list
        extracted_keywords += keyword
    return extracted_keywords

# Fix the "keywords" column by converting cells from a list to a string
df_reviews_List['keywords'] = df_reviews_List['keywords'].apply(extract_keywords)
```

### Access The Movie Database API

```python
# Prepare The Movie Database query
tmdb_url = "https://api.themoviedb.org/3/search/movie?query="
tmdb_key_string = "&api_key=" + tmdb_api_key

tmdb_api_url='https://api.themoviedb.org/3/search/movie'
#https://api.themoviedb.org/3/movie/343611?api_key=XXXXXXXXX

def getMovieByName(tmdb_title):
    tmdb_qry_params = {
        "query": tmdb_title,
        "api_key": tmdb_api_key 
    }
    #print (tmdb_qry_params)
    tmdb_response = requests.get(tmdb_api_url,params=tmdb_qry_params)
    return tmdb_response
def getMovieById(id):
    tmdb_qry_params = {
        "api_key": tmdb_api_key 
    }
    #print (tmdb_qry_params)
    tmdb_response = requests.get('https://api.themoviedb.org/3/movie/{}'.format(id),params=tmdb_qry_params)
    return tmdb_response
```
